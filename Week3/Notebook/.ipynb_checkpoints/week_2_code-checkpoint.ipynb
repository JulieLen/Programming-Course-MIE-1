## A Working Example
Consider a gambler who has $\$100$. Our friendly gambler comes across another, very rich gambler who makes him a proposition: they will flip a (possibly biased) coin and if it is heads, the gambler wins $\$1$, but if it is tails he loses $\$1$. The stranger is even kind enough to tell him the probability of a heads is $p$.

This is one of the oldest problems in probability theory, known as the gambler's ruin (well, a spin-off of that). We can ask many questions of this problem, such as "What is the probability of doubling your money before you go broke?" or "When is an optimal time to stop betting?".

Although many of these questions have analytical answers, I'm assuming you have absolutely no idea about probability, so we'll instead do some experiments and learn about R at the same time!

## Random Numbers

The next thing we need is to generate the outcome for our gambler's bet. To do this, we will use a <a href="https://en.wikipedia.org/wiki/Pseudorandom_number_generator" target="_blank">pseudo-random number generator</a>.

It is impossible for a computer to generate a truly random number, so this is the next best thing.

There are several ways to generate random numbers, the first being to use an inverse cumulative density function, or CDF.

First, one generates a uniform random variable using `runif(n)`, where `n` is the dimension:

````r
r = runif(1)
r
````

Then, one inputs this value into the inverse of the CDF. What is the CDF in our case....?

If for each gamble, the outcome is a random variable $X$ equal to $1$ for a win and $0$ for a loss, we can define the pmf $f(x) = \mathbb{P}(X = x)$ as
$$
f(x)=
\begin{cases}
    p,& \text{if } x = 1\\
    1 - p,              & \text{if } x = 0
\end{cases}
$$

Then the CDF $F(x) = \mathbb{P}(X \leq x)$ can be defined in a similar way:
$$
F(x)=
\begin{cases}
     1 - p,& \text{if } x < 1\\
     p,& \text{otherwise}
\end{cases}
$$

Ok, but how the heck is this helping us generate random variables?! Well, think of our uniform random number `r` as a probability. Then this corresponds to a single element of the sample space. Simply invert the cdf to see what the corresponding value of $X$ is (**draw this on the board. Hint: Draw the discrete CDF... the generalized inverse connects the dots!**).

This doesn't necessarily help... the sample space of $X$ is countable, so the inverse $F^{-1}(u)$ is not one-to-one. We need the generalized inverse $F^{-1}(u)= \inf\{ x : F(x)\geq u \}$

This is called the **Probability Integral Transform** (see Casella and Burger, Example 5.6.3).

So, let's generate one random draw from our gambler's bet:

````r
p <- 0.5
u <- runif(1)
if(u <= 1 - p){
    x <- 0
} else {
    x <- 1
}
x
````

## If Statements
An `if` statement is conditioning statement which returns a **boolean**, or a True/False outcome.

The syntax for `if` statements differs across languages, but the logic is always the same. If the condition is true, the code is executed and if the statement is false the code is skipped.

One can also include an `else` statement, which is executed in the case the condition is not satisfied.  

The key is the **boolean**. This is a function that outputs either True or False depending on the input.

Let's see a simple example:

````r
number <- 1

number == 1
number == 2
````

You should be able to figure out what `==` means pretty easily, but there are plenty of camparison operators one can use:

    !x           % Negation
    x & y        % Elementwise intersection
    x && y       % Global intersection (output one boolean)
    x | y        % Elementwise union
    x || y       % Global union (output one boolean)
    xor(x, y)    % Elementwise exclusive union
    isTrue(x)    % Input boolean, output boolean
    
You should try to become familiar with how to structure your programs under True/False conditions and with the different types of logical operations.

A side point is that one can nest if statements, and the syntax is:

````r
number <- 4
if(number == 1){
    print("I'm a one!")
}else if(number == 2){
    print("I'm a two!")
}else if(number == 3){
    print("I'm a three!")
}else{
    print("I can't count that high!")
}
````

````r
vector <- c(1.0, 2.0, 3.0)
matrix <- matrix(c(vector, vector), nrow=3, ncol=2)
array <- array(c(vector, vector, vector, vector), dim=c(3,2,2))

print("Dimension")

print(dim(vector))
print(dim(matrix))
print(dim(array))

print("Length")

print(length(vector))
print(length(matrix))
print(length(array))
````

Ok, so now we understand our `if` statement, so let's use our random number generator to let our gambler play one round. I'll re-define our variables just so things are clear:

````r
w <- 100 # Initial wealth
p <- 0.5 # Probability of a win


sprintf("You started with $%s", w)

# A random draw
u <- runif(1)
if(u <= 1 - p){
    print("You lose!")
    x <- 0
    w <- w - 1
} else {
    print("You win!")
    x <- 1
    w <- w + 1
}
sprintf("Now you've got $%s.", w)
````

We've done it! Now our gambler can play to his heart's content, but there's something wrong with this picture...

How can we make it so the gambler can play multiple times with the same wealth?

## For loops
Now that we can gamble once, we can gamble many times. We'll use a classic construct known as a `for` loop. The idea is to loop over items in some set, just like in a mathematics formula which reads $\forall i \in$...

This is maybe easiest seen in an example:

````r
I <- 10
for(i in 1:I){
    print(sprintf("i = %s", i))
}
````

As you can see, we are looping through all the values of `i` less than or equal to `I`. We can also loop over other objects:

````r
X <- c(1:10)
for(x in X){
    print(sprintf("x = %s", x))
} 
````

We can also use a construct called a `while` loop, which will continue looping until some condition is no longer met. The difference between `for` and `while` loops is that one is **definite** while the other is **indefinite**. One knows exactly when a **definite** loop will end, while one isn't necessarily sure when an **indefinite** loop will end.

Here's an example of a while loop just like our for loops:

````r
I <- 10
i <- 0
while(i <= I){
    print(sprintf("i = %s", i))
    i = i + 1
}
````

Notice that without the last line, this loop would run forever.

Now let's say we would like to allow our gambler to play up until some time `T`. We'll simply define our `T` and re-use our code from before:

````r
w <- 100 # Initial wealth
p <- 0.5 # Probability of a win
T <- 10  # Terminal time

sprintf("You started with $%s", w)

for(t in 1:T){
    # A random draw
    u <- runif(1)
    if(u <= 1 - p){
        print("You lose!")
        x <- 0
        w <- w - 1
    } else {
        print("You win!")
        x <- 1
        w <- w + 1
    }
    print(sprintf("Now you've got $%s.", w))
    # NOTE: The below doesnt work in the notebook...
    # print ("Press [enter] to continue")
    # number <- scan(n=1)
}
````

## Using Vectors to Track Wealth and Wins
That's great, but we are losing a lot of data through time. What if we want to save the agents wins and losses or wealth through time? We can do this using vectors.

The simplest approach is to create some vectors to fill and loop over the time periods, using the iteration index as the vector index:

````r
w0 <- 100 # Initial wealth
p <- 0.5 # Probability of a win
T <- 10  # Terminal time

sprintf("You started with $%s", w0)

# Create empty objects to fill
w <- rep(0, T+1)
x <- rep(0, T)

# Fill initial wealth
w[1] = w0

for(t in 1:T){
    # A random draw
    u <- runif(1)
    if(u <= 1 - p){
        print("You lose!")
        x[t] <- 0
        w[t+1] <- w[t] - 1
    } else {
        print("You win!")
        x[t] <- 1
        w[t+1] <- w[t] + 1
    }
    print(sprintf("Now you've got $%s.", w[t+1]))
    # NOTE: The below doesnt work in the notebook...
    # print ("Press [enter] to continue")
    # number <- scan(n=1)
}
````

Great! Now we can study the outcomes of our gambler's gambles, for example, we can look at the average number of wins or losses, the average time spent with profits versus losses, etc.

Another thing we can do is create a simple plot of the outcomes.

## Basic Plotting
It's nice to be able to visualize your data, so for now we'll cover some simple, one-dimensional plotting. Later we'll do something more complex.

Creating a plot is as simple as one line, where you give the `plot()` function the x values and corresponding y values:

````r
x <- 1:10
y <- x^2
plot(x, y)
````

If you would like to include a line, simply define the `type`:

````r
plot(x, y, type='l')
````

Now let's look at our gambler's wealth over time. I'll supress the verbose output so we can look at longer periods:

````r
w0 <- 100 # Initial wealth
p <- 0.5 # Probability of a win
T <- 10000  # Terminal time

# Create empty objects to fill
w <- rep(0, T+1)
x <- rep(0, T)

# Fill initial wealth
w[1] = w0

for(t in 1:T){
    # A random draw
    u <- runif(1)
    if(u <= 1 - p){
        x[t] <- 0
        w[t+1] <- w[t] - 1
    } else {
        x[t] <- 1
        w[t+1] <- w[t] + 1
    }
}
time = 0:T
plot(time, w, type='l')
````
